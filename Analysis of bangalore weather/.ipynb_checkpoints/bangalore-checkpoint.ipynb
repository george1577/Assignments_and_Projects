## Data Analysis with pandas

We are going to examine data from the [Bangalore Open Data Repository](https://github.com/openbangalore/bangalore). Bangalore is the 3rd most populous city in India and widely regarded as the Silicon Valley of India.

### Task 1 - Load Data

- Read in the data to pandas
- Check the shape of the dataframe
- Check the summary statistics of the dataframe
- Change the row index to the "Year" column

```python
import pandas as pd
import numpy as np
banglore = pd.read_csv('bangalore_temparature.tsv', delimiter = '\t')
#Check the shape of the dataframe
print(banglore.shape)
#Check the summary statistics of the dataframe
print(banglore.describe())
#Change the row index to the "Year" column
#banglore['Year'] = range(0, len(banglore))
banglore = banglore.reset_index()
banglore = banglore.set_index(banglore['Year'])

```

### Task 2 - Initial Data Analysis

- (Chart) How does temperature vary over the year (X-axis is months)?
- (Chart) How does temperature vary over the years (X-axis is years)?
- Which months had the highest and lowest temperatures in 1960?
- What were the highest, lowest and mean values in 1960?
- Where were the highest, lowest and mean values in an arbitray year (hint: write a function)?
- Which months had the highest and lowest gains in temperature?
- Make a histogram, KDE plot and bar plot of the gains.

```python
import matplotlib.pyplot as plt
banglore_copy = banglore.copy().drop(['index','Year'], axis=1)
df = banglore_copy.mean()
plt.subplot(1,2,1)
plt.plot(list(range(1,13)), list(df))
plt.xlabel('Month')
plt.ylabel('Avg Temperature')
plt.xticks(list(range(1,13)))

plt.subplot(1,2,2)
df2 = banglore_copy.mean(axis = 1)
plt.xlabel('Year')
plt.ylabel('Avg Temperature')
plt.plot(df2)
plt.subplots_adjust(wspace=.8)

def find_max_min_mean(year):
      h = banglore_copy.loc[year].max()
      l = banglore_copy.loc[year].min()
      avg = banglore_copy.loc[year].mean()
      return h, l, avg
a1, b1, c1 = find_max_min_mean(1960)
x, y = banglore_copy.loc[1960].idxmax(), banglore_copy.loc[1960].idxmin() 
print("In year 1960, highest temp is {}, lowest temp is {} and mean temp is {}".format(a1,b1,c1))
print('In year 1960, highest temp month is {} and lowest temp month is {}'.format(x,y))

#Which months had the highest and lowest gains in temperature?
print('Highest gains: ' + banglore_copy.loc[1960].diff().dropna().idxmax())
print('Lowest gains: ' + banglore_copy.loc[1960].diff().dropna().idxmin())

#Make a histogram, KDE plot and bar plot of the gains.
banglore_diff = banglore_copy.diff().dropna()
banglore_diff.plot(kind = 'hist', figsize = (18,10))
banglore_diff.plot(kind = 'bar', figsize = (18,10))
banglore_diff.plot(kind = 'kde', figsize = (18,10))

```

### Task 3 - Visualization

matplotlib
- Make a histogram for a particular month.
- Compare distributions for 2 months (histograms on subplots).
- Compare two years in the same histogram (use color coding).
- Define a function to compare two months in the same frame.
- Create a box plot for 1 month.
- Create a box plot for 2 months on 1 figure.

seaborn
- Make a KDE plot of one month.
- Make a function that compares the KDE plots of 2 months.
- Compare the histogram and KDE plots of 2 months on the same figure.
- Create a Facetgrid version of the KDE plots. Loop through a months array.
- Create a violin plot for 1 month.
- Create a violin plot for 2 months on 1 figure.

```python
#Make a histogram for a particular month.
banglore_diff['Jan'].plot(kind = 'hist', figsize = (18,10))

#Compare distributions for 2 months (histograms on subplots).
banglore_diff[['Jan', 'Jun']].plot(kind = 'hist', figsize = (18,10), alpha = 0.5)

#Compare two years in the same histogram (use color coding).
plt.figure()
y1950 = banglore_diff.loc[1950].plot(kind = 'hist', figsize = (18,10), alpha = 0.5)
y1950.legend(bbox_to_anchor=(1.05, 1), loc=2, ncol=5)
y1960 = banglore_diff.loc[1960].plot(kind = 'hist', figsize = (18,10), alpha = 0.5)
y1960.legend(bbox_to_anchor=(1.05, 1), loc=2, ncol=5)

#Create a box plot for 1 month.
plt.figure() 
banglore_diff['Jan'].plot(kind = 'box', figsize = (18,10))

#Create a box plot for 2 months on 1 figure.
plt.figure()
banglore_diff[['Jan', 'Jul']].plot(kind = 'box', figsize = (18,10))

#Create a function that compares 2 months via boxplot.
def compare_box(m1, m2):
    banglore_diff[[m1,m2]].plot(kind = 'box', figsize = (18,10))

#Make a KDE plot of one month.
plt.figure()
banglore_diff['Jan'].plot(kind = 'kde', figsize = (18,10))
     
#Make a function that compares the KDE plots of 2 months.
def compare_kde(m1, m2):
    banglore_diff[[m1,m2]].plot(kind = 'kde', figsize = (18,10))
compare_kde('Jan', 'Jul')    
    
#Compare the histogram and KDE plots of 2 months on the same figure.
import seaborn as sns
plt.figure()
sns.distplot(banglore_copy['Jan'], hist = True, kde = True, bins = 15)

#Create a Facetgrid version of the KDE plots. Loop through a months array.

#Create a violin plot for 1 month.
plt.figure()
sns.violinplot(banglore_copy['Jan'])

#Create a violin plot for 2 months on 1 figure.
plt.figure()
sns.violinplot(banglore_copy['Jan'])
sns.violinplot(banglore_copy['Jul'])

#Create a function that compares 2 months via violin plot.
def compare_violin(m1, m2):
    plt.figure()
    sns.violinplot(banglore_copy[m1])
    sns.violinplot(banglore_copy[m2])



```
